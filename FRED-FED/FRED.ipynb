```python
from pandas_datareader.data import DataReader
from datetime import date

# Define parameters
series_code = 'DGS2'  # 10-year Treasury Rate
data_source = 'fred'  # FED Economic Data Service
start_date = date(1962, 1, 1)

try:
    # Fetch data from FRED
    data = DataReader(series_code, data_source, start_date)
    
    # Display basic information about the data
    print("Data Information:")
    print(data.info())
    
except Exception as e:
    # Handle any errors that might occur during data retrieval
    print("Error:", e)

import matplotlib.pyplot as plt
# Rename the columns for clarity
data = data.rename(columns={'Close': 'Stock Close', 'DCOILWTICO': '2y Treasury'})

# Plot the data
data.plot(title='2y Treasury')
plt.xlabel('Date')
plt.ylabel('Rate')
plt.show()
    
```
