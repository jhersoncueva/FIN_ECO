```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python

data = yf.download("META", start="2020-01-01", end="2024-04-12")
print(data.head())
```

```python
# Convertir el ndice a tipo datetime
data.index = pd.to_datetime(data.index)

# Graficar los precios de cierre
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], label='Precio de cierre', color='blue')

# Configurar el eje x como fecha
plt.xlabel('Fecha')
plt.ylabel('Precio de cierre')
plt.title('Grfico de precios de cierre de AAPL')
plt.legend()
# plt.grid(True)

# Mostrar la grfica
plt.show()
```

```python
# Convertir el ndice a tipo datetime
data.index = pd.to_datetime(data.index)

# Graficar los precios de cierre
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Volume'], label='Volumen', color='blue')
# Configurar el eje x como fecha
plt.xlabel('Fecha')
plt.ylabel('Volumen')
plt.title('Grfico de volumen negociado de AAPL')
plt.legend()
# plt.grid(True)

# Mostrar la grfica
plt.show()
```
