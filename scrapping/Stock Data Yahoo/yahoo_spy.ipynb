```python
import yfinance as yf
import pandas as pd
import datetime as dt

# Fechas de inicio y fin
start = dt.datetime(2020, 1, 7)
end = dt.datetime.now()

#US
sp500 = yf.download('^GSPC', start=start, end=end)[['Close']]  # S&P 500
nasdaq = yf.download('^IXIC', start=start, end=end)[['Close']]  # Nasdaq Composite
dowjones = yf.download('^DJI', start=start, end=end)[['Close']]  # Dow Jones Industrial Average
russell2000 = yf.download('^RUT', start=start, end=end)[['Close']]  # Russell 2000

sp500.rename(columns={'Close': 'S&P 500'}, inplace=True)
nasdaq.rename(columns={'Close': 'Nasdaq'}, inplace=True)
dowjones.rename(columns={'Close': 'Dow Jones'}, inplace=True)
russell2000.rename(columns={'Close': 'Russell 2000'}, inplace=True)


# Europa
ftse = yf.download('^FTSE', start=start, end=end)[['Close']]
dax = yf.download('^GDAXI', start=start, end=end)[['Close']]
cac = yf.download('^FCHI', start=start, end=end)[['Close']]
stoxx = yf.download('^STOXX50E', start=start, end=end)[['Close']]


ftse.rename(columns={'Close': 'FTSE 100'}, inplace=True)
dax.rename(columns={'Close': 'DAX'}, inplace=True)
cac.rename(columns={'Close': 'CAC 40'}, inplace=True)
stoxx.rename(columns={'Close': 'Euro Stoxx 50'}, inplace=True)

# Asia
nikkei = yf.download('^N225', start=start, end=end)[['Close']]
hangseng = yf.download('^HSI', start=start, end=end)[['Close']]
shanghai = yf.download('000001.SS', start=start, end=end)[['Close']]
kospi = yf.download('^KS11', start=start, end=end)[['Close']]

nikkei.rename(columns={'Close': 'Nikkei 225'}, inplace=True)
hangseng.rename(columns={'Close': 'Hang Seng'}, inplace=True)
shanghai.rename(columns={'Close': 'Shanghai Composite'}, inplace=True)
kospi.rename(columns={'Close': 'Kospi'}, inplace=True)

# Latinoamérica
bovespa = yf.download('^BVSP', start=start, end=end)[['Close']]
ipc_mexico = yf.download('^MXX', start=start, end=end)[['Close']]
merval = yf.download('^MERV', start=start, end=end)[['Close']]

bovespa.rename(columns={'Close': 'Bovespa'}, inplace=True)
ipc_mexico.rename(columns={'Close': 'IPC México'}, inplace=True)
merval.rename(columns={'Close': 'Merval'}, inplace=True)


#Perú
sp_bvl = yf.download('^SPBLPGPT', start=start, end=end)[['Close']]
sp_bvl.rename(columns={'Close': 'S&P/BVL General'}, inplace=True)

EPU = yf.download('EPU', start=start, end=end)[['Close']]
EPU.rename(columns={'Close': 'MSCI Peru ETF'}, inplace=True)

# Concatenar todos los índices en un DataFrame
indices_data = pd.concat([sp500, nasdaq, dowjones, russell2000,ftse, dax, cac, stoxx, nikkei, hangseng, shanghai, kospi, bovespa, ipc_mexico, merval,sp_bvl,EPU], axis=1)
# indices_data = pd.concat([indices_data, sp_bvl], axis=1)

# Normalizar los datos a 100 tomando el valor del primer día
# indices_data = indices_data / indices_data.iloc[0] * 100

indices_data.to_excel("indices_data_US.xlsx", sheet_name="indices_data_US")
# Mostrar las primeras filas del DataFrame
indices_data.head()
```

```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt

# Definir el rango de fechas
start = datetime.datetime(2000, 1, 1)
end = datetime.datetime(2024, 9, 27)

# Obtener los datos de inflación (CPI), core inflation (Core CPI), y PCE de la FRED
data = web.DataReader(['CPIAUCSL', 'CPILFESL'], 'fred', start, end)


# Calcular la tasa de inflación interanual (YoY) para cada serie
inflation_yoy = data.pct_change(periods=12) * 100
print(inflation_yoy.tail(4))

inflation_mom = data.pct_change(periods=1) * 100
inflation_mom.tail(4)

inflation_yoy.to_excel("data_US.xlsx",sheet_name="data_US")
```

```python
# Graficar las tasas de inflación interanual
plt.figure(figsize=(10, 6))
plt.plot(inflation_yoy['CPIAUCSL'], label='Inflación Total (CPI)')
plt.plot(inflation_yoy['CPILFESL'], label='Core CPI (Inflación Subyacente)')

# Añadir títulos y etiquetas
plt.title('Tasa de Inflación YoY en EE.UU. (2000-2024)')
plt.xlabel('Fecha')
plt.ylabel('Inflación YoY (%)')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
import pandas_datareader.data as web
import pandas as pd
import matplotlib.pyplot as plt
import datetime

# Definir el rango de fechas
start = datetime.datetime(2014, 1, 1)
end = datetime.datetime(2024, 9, 30)

# Obtener los datos del PCE (Personal Consumption Expenditures) y Core PCE de la economía de EE.UU. de la FRED
pce = web.DataReader(['PCEPI', 'PCEPILFE'], 'fred', start, end)


pce["var_PCEPI"] = pd.DataFrame(pce["PCEPI"].pct_change(12)*100)
pce["var_PCEPOILFE"] = pd.DataFrame(pce["PCEPILFE"].pct_change(12)*100)

pce = pce.dropna()

pce.to_excel("data_ECONOMICA.xlsx",sheet_name="data_ECONOMICA")

plt.figure(figsize=(10, 6))
plt.plot(pce["var_PCEPI"],label='PCE (Personal Consumption Expenditure)')
plt.plot(pce["var_PCEPOILFE"],label='Core PCE (Core Personal Consumption Expenditure)')
plt.legend()
plt.show()
```
