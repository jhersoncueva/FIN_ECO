```python
data = yf.download("META", start="2020-01-01", end="2024-04-12")
print(data.head())
```

```python
# Convertir el índice a tipo datetime
data.index = pd.to_datetime(data.index)

# Graficar los precios de cierre
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], label='Precio de cierre', color='blue')

# Configurar el eje x como fecha
plt.xlabel('Fecha')
plt.ylabel('Precio de cierre')
plt.title('Gráfico de precios de cierre de AAPL')
plt.legend()
# plt.grid(True)
plt.savefig('grafico_precios_cierre_AAPL_2020.pdf')
# Mostrar la gráfica
plt.show()
```

```python
# Convertir el índice a tipo datetime
data.index = pd.to_datetime(data.index)

# Graficar los precios de cierre
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Volume'], label='Volumen', color='blue')
# Configurar el eje x como fecha
plt.xlabel('Fecha')
plt.ylabel('Volumen')
plt.title('Gráfico de volumen negociado de AAPL')
plt.legend()
# plt.grid(True)

plt.savefig('grafico_precios_cierre_AAPL_2020.pdf')
# Mostrar la gráfica
plt.show()
```
